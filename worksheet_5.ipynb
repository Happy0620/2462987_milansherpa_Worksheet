```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("student.csv")

print(data.head())
print(data.describe())
```

```python
def linear_regression(x, y):
    x_mean, y_mean = np.mean(x), np.mean(y)
    num = np.sum((x - x_mean) * (y - y_mean))
    den = np.sum((x - x_mean) ** 2)
    beta1 = num / den
    beta0 = y_mean - beta1 * x_mean
    return beta0, beta1
```

```python
X = data["Math"].values
y = data["Writing"].values

beta0, beta1 = linear_regression(X, y)
print(f"Intercept (β0): {beta0:.2f}, Slope (β1): {beta1:.2f}")
```

```python
def predict(x, beta0, beta1):
    return beta0 + beta1 * x

y_pred = predict(X, beta0, beta1)
```

```python
mse = np.mean((y - y_pred) ** 2)
r2 = 1 - (np.sum((y - y_pred) ** 2) / np.sum((y - np.mean(y)) ** 2))

print(f"MSE: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y_pred, color="red", label="Regression Line")
plt.xlabel("Math Scores")
plt.ylabel("Writing Scores")
plt.title("Linear Regression: Math vs Writing")
plt.legend()
plt.show()
```
