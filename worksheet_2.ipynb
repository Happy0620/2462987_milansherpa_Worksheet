```python
numeric_cols = ['Age', 'Height', 'Weight', 'BMI', 'Temperature', 'Heart Rate',
                'Blood Pressure', 'Cholesterol']
for col in numeric_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

categorical_cols = ['Gender', 'Blood Type', 'Diabetes', 'Smoking']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")


```

```python
print("Duplicates:", df.duplicated().sum())
df = df.drop_duplicates()

```

```python

titanic = pd.read_csv("Titanic-Dataset.csv")
titanic.head()
```

```python
subset = titanic[['Name','Pclass','Sex','Age','Fare','Survived']]
first_class = subset[subset['Pclass'] == 1]

print("Fare mean:", first_class['Fare'].mean())
print("Fare median:", first_class['Fare'].median())
print("Fare max:", first_class['Fare'].max())
print("Fare min:", first_class['Fare'].min())
```

```python
subset['Age'].isnull().sum()
```

```python
subset_no_null_age = subset.dropna(subset=['Age'])
subset_no_null_age.head()
```

```python
embarked_encoded = pd.get_dummies(titanic['Embarked'], prefix='Embarked')
titanic = pd.concat([titanic, embarked_encoded], axis=1)
titanic = titanic.drop(columns=['Embarked'])
titanic.head()
```

```python
titanic.groupby('Sex')['Survived'].mean()
```

```python
titanic.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
sns.barplot(x='Embarked', y='Survived', data=pd.read_csv("Titanic-Dataset.csv"))
plt.title("Survival Rate by Port of Embarkation")
plt.show()
```

```python
titanic['AgeGroup'] = pd.qcut(titanic['Age'], 5)
titanic.groupby(['Pclass','AgeGroup'])['Survived'].mean()
```

```python
sns.catplot(x='Pclass', y='Survived', hue='AgeGroup', data=titanic, kind='bar')
plt.title("Survival Rate by Age Group and Class")
plt.show()
```
