```python
time_data = [
    # (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0)
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]
```

```python
low, moderate, high = [], [], []

for study, entertainment, sleep in time_data:
    if study < 3:
        low.append(study)
    elif 3 <= study <= 5:
        moderate.append(study)
    else:
        high.append(study)

print("Low:", low)
print("Moderate:", moderate)
print("High:", high)
```

```python
print("Low count:", len(low))
print("Moderate count:", len(moderate))
print("High count:", len(high))
```

```python
study_minutes = [int(study * 60) for study, _, _ in time_data]
print("Study minutes:", study_minutes)
```

```python
study_hours = [s for s, _, _ in time_data]
entertainment_hours = [e for _, e, _ in time_data]
sleep_hours = [sl for _, _, sl in time_data]

def average(nums):
    return sum(nums) / len(nums)

print("Avg study:", round(average(study_hours), 2))
print("Avg entertainment:", round(average(entertainment_hours), 2))
print("Avg sleep:", round(average(sleep_hours), 2))

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.scatter(study_hours, sleep_hours, c='steelblue', edgecolor='black')
plt.title("Study vs Sleep (Daily)")
plt.xlabel("Study hours")
plt.ylabel("Sleep hours")
plt.grid(True, alpha=0.3)
plt.show()

```

```python
def factorial(n):
    """
    Returns n! for non-negative integer n.
    Raises ValueError for negative inputs.
    """
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(5))  

```
