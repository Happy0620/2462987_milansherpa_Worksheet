matrix operation

```python
A = np.array([[3,4],[7,8]])
A_inv = np.linalg.inv(A)
print(np.dot(A, A_inv))
```

```python
B = np.array([[5,3],[2,1]])
print("AB:\n", np.dot(A,B))
print("BA:\n", np.dot(B,A))
```

```python
AB = np.dot(A,B)
print("(AB)^T:\n", AB.T)
print("B^T A^T:\n", np.dot(B.T, A.T))
```

```python
coeff = np.array([[3,1,-1],[2,-3,1],[1,-1,2]])
const = np.array([9,-1,-3])
solution_inv = np.dot(np.linalg.inv(coeff), const)
print(solution_inv)
```

```python
solution_solve = np.linalg.solve(coeff, const)
print(solution_solve)
```

```python
import time 
listA = [[i for i in range(1000)] for _ in range(1000)]
listB = [[i for i in range(1000)] for _ in range(1000)]
start = time.time()
result = [[sum(a*b for a,b in zip(row,col)) for col in zip(*listB)] for row in listA]
print("List matrix multiplication time:", time.time()-start)
```

```python
arrA = np.array(listA)
arrB = np.array(listB)
start = time.time()
result_np = np.dot(arrA, arrB)
print("NumPy matrix multiplication time:", time.time()-start)
```
